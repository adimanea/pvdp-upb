##  Descriere problema+graficul functiei alese

### Problema consta in analizarea unei fuinctii de doua variabile(R^2->R), aflarea extremelor libere ale acesteia si realizarea interpretarii grefice . In scop de exemplificare vom utiliza functie urmatoare si ii vom realiza graficul: *f:R^2->R ,   f(x,y)=x^2-y^3-6cosxsinx*

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

x = np.linspace(-10,10,300);
y = np.linspace(-10,10,300);

x,y=np.meshgrid(x,y)
z=x**2-y**2-6*np.cos(x)*np.sin(y)


#GRAFIC WIREFRAME---
#ax.plot_wireframe(x,y,z, rstride=6, cstride=6)

#GRAFIC SUPRAFATA---
surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm,linewidth=1, antialiased=False)


plt.show()

```

## Rezolvare(PB1)

### Aflarea extremelor functiei de doua variebile definita astfel:  *f:R^2->R ,    f(x,y)=3xy^2-x^3-15x-36y+9*

### Derivatele patiale 

```python
f=input("Enter your value: ")
x = var('x')
y = var('y')
fx=diff(f, x)
show(fx==0)
fy=diff(f, y)
show(fy==0)
```

### Rezolvarea sistemului

```python
show(solve([fx==0, fy==0],x,y))
```

### ==>punctele critice sunt **A(2,3)** si **B(-2,-3)**

### Matricea hessiana

```python
fxx=diff(fx,x)
fxy=diff(fx,y)
fyx=diff(fy,x)
fyy=diff(fy,y)
print('H=')
show(matrix([[fxx,fxy],[fyx,fyy]]))
```

### A(2,3)

```python
var ('L1')
print('Ha=')
M=matrix([[fxx.subs(x = 2),fxy.subs(y = 3)],[fyx.subs(y = 3),fyy.subs(x = 2)]])
show(M)
M1=L1*identity_matrix(2)
print('Pm=')
show(M-M1)
print(M.charpoly('L1'))
solve(L1^2 - 468==0,L1)
```

### Valorile proprii generate de pct. A au produsul < 0 ==> A nu este extrem
### B(-2,-3)

```python
var ('L2')
print('Hb=')
N=matrix([[fxx.subs(x = -2),fxy.subs(y = -3)],[fyx.subs(y = -3),fyy.subs(x = -2)]])
show(N)
N1=L2*identity_matrix(2)
print('Pn=')
show(N-N1)
print(N.charpoly('L2'))
solve(L2^2 - 468,L2)
```

### Valorile proprii generate de pct. B au produsul < 0 ==> B nu este extrem
