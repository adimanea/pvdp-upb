<font size = "4">1. In prima faza desenez graficul si reprezint ecuatiile date pentru a putea vizualiza domeniul.</font>

```python
import numpy as np
import matplotlib.pyplot as plt
```

<font size = "4.5">-Reprezint curba data de ecuatia y=x<sup>2</sup>  
Aceasta va fi o parabola cu varful in jos si in origine si fixata in origine.</font>

```python
# il scriu pe x ca apartinand intervalului [-4,4], avand 1000 de puncte intre -4 si 4
x = np.linspace(-4,4, 1000)

# scriu ecuatia
y = x**2

# afisez graficul
plt.plot(x, y, color = 'blue')
plt.axvline(x=0, color = 'black')
plt.grid(color = 'grey',alpha = 0.3, linestyle='-', linewidth=2)

# legenda
plt.legend(['y = x^2', 'axa y'])
plt.show()
```

<font size = "4.5">-Reprezint curba data de ecuatia x = y<sup>2</sup>, care va fi o parabola cu varful inspre stanga in origine, de-a lungul dreptei Ox. Pentru a-mi fi mai usor am sa reprezint ecuatiile y = $\sqrt{x}$ si y = -$\sqrt{x}$</font>

```python
x = np.linspace(0, 4, 1000)
y1 = np.sqrt(x)
y2 = -np.sqrt(x)
plt.plot(x, y1, color = 'red')
plt.axhline(y=0, color = 'black')
plt.plot(x, y2, color = 'red')
plt.grid(color = 'grey', alpha = 0.3, linestyle='-', linewidth=2)
plt.legend(['x = y^2', 'axa y'])
plt.show()
```

<font size="4.5">-Intersectez cele doua grafice</font>

```python
# scriu cele doua ecuatii date
x = np.linspace(0, 1, 100)
y1 = x**2
y2 = np.sqrt(x)

# proiectiile punctului de intersectie (1, 1)
xx = [1, 1]
yy = [1, 0]

# punctele pentru a-mi trasa axele x si y
x_axa = [0, 0]
y_axa = [0, 2]
xx_axa = [0, 2]
yy_axa = [0, 0]

# reprezint graficul si hasurez aria cuprinsa de cele doua curbe
plt.plot(x, y1, x, y2)
plt.fill_between(x, y1, y2, where = (x<1), color = 'cyan', alpha = 0.6)

# reprezint axa y
plt.plot(x_axa, y_axa, 'g-^')

# reprezint axa x
plt.plot(xx_axa, yy_axa, 'r->')

# reprezint punctuele de intersectie ale celor 2 curbe
plt.plot((0,1), 'o', color = 'black')

# unesc punctul de intersectie su proiectia pe axa x
plt.plot(xx, yy, '--', color = 'black')

# unesc punctul de intersectie cu axa y
plt.plot((1, 1), '--', color = 'black')

# denumesc punctele de intersectie
plt.annotate("Origine", (0, 0))
plt.annotate("A", (1, 1))

# scriu ce reprezinta fiecare linie din plot
plt.legend(['y = x^2', 'y^2 = x', 'axa y', 'axa x'])

plt.grid()
plt.show()
```

</font>

<font size = "4.5">-Integrala construita arata astfel:
    $$\int_0^1 \int_{x^2}^{âˆšx} (x+y) \,dxdy$$</font>

<font size = "4">Primul pas: rezolvare simbolica.</font>

```python
from sympy import Symbol
from sympy import *
x = Symbol('x')
y = Symbol('y')
print("I =")
integrate((x+y), (y, x*x, sqrt(x)), x)
```

<font size = "4">Al doilea pas: rezolvare numerica.</font>

```python
print("I =", integrate((x+y), (y, x*x, sqrt(x)), (x, 0, 1)))
```
